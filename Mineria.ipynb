The Basics of NumPy Arrays

```python
import numpy as np
np.random.seed(0)  #semilla random

x1 = np.random.randint(10, size=6)  # array de una dimension
x2 = np.random.randint(10, size=(3, 4))  # array de dos dimensiones
x3 = np.random.randint(10, size=(3, 4, 5))  # array de tres dimensiones
```

```python
print("x3 ndim: ", x3.ndim) # la propiedad ndim nos dice el numero de dimensiones
print("x3 shape:", x3.shape) # shape dimesiones
print("x3 size: ", x3.size) # tamaño se multiplican los numeros ej  size=(3, 4, 5)
```

```python
print("dtype:", x3.dtype) #tipo de dato
```

```python
print("itemsize:", x3.itemsize, "bytes") #tamaño de bytes
print("nbytes:", x3.nbytes, "bytes") # numero de bytes
```

#### Estadistica Basica

Media (Mean)

```python
data = np.array([1, 2, 3, 4, 5])
media = np.mean(data)
print(media)
```

Promedio (Average) y Promedio Ponderado

```python
data = np.array([1, 2, 3, 4, 5])
weights = np.array([1, 2, 3, 4, 5])
promedio = np.average(data)
promedio_ponderado = np.average(data, weights=weights)
```

```python
print(promedio)
print(promedio_ponderado)
```

Mediana

```python
mediana = np.median(data)
```

```python
print(mediana)
```

Varianza y Desviación Estándar

```python
varianza = np.var(data)
desviacion = np.std(data)
```

```python
print(varianza)
print(desviacion)
```

Generación de Datos Aleatorios

```python
muestra = np.random.normal(0, 1, 1000)
print(np.mean(muestra), np.std(muestra))
```

Correlación

```python
corr = np.corrcoef([1,2,3], [1,5,7])
```
