```python
import matplotlib.pyplot as plt

plot_data_english = english_color_percentage_df.sort_values(by='english_percentage', ascending=False)

plt.figure(figsize=(10, 6))
plt.barh(plot_data_english['color'], plot_data_english['english_percentage'], color='blue')
plt.title('Porcentaje de Colores en Inglés')
plt.xlabel('Porcentaje (%)')
plt.ylabel('Color')
plt.gca().invert_yaxis()
plt.show()
```

```python

plt.figure(figsize=(10, 8))
plt.scatter(x=english_spanish_correlation['english_percentage'], y=english_spanish_correlation['spanish_percentage'], color='blue', s=100)
for i, row in english_spanish_correlation.iterrows():
    plt.text(row['english_percentage'] + 0.5, row['spanish_percentage'], row['color'], horizontalalignment='left', size='small', color='black', weight='semibold')
plt.title('Correlación de Porcentajes: Inglés vs Español ')
plt.xlabel('Porcentaje de Colores en Inglés')
plt.ylabel('Porcentaje de Colores en Español')
plt.grid(True)
plt.show()
```
