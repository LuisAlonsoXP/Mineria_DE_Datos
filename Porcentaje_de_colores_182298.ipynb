```python
import matplotlib.pyplot as plt
import seaborn as sns

plt.figure(figsize=(10, 8))
sns.scatterplot(x='english_percentage', y='spanish_percentage', data=english_spanish_correlation, color='blue', s=100)
for i, row in english_spanish_correlation.iterrows():
    plt.text(row['english_percentage'] + 0.5, row['spanish_percentage'], row['color'], horizontalalignment='left', size='small', color='black', weight='semibold')
plt.title('Correlación de Porcentajes: Inglés vs Español')
plt.xlabel('Porcentaje de Colores en Inglés')
plt.ylabel('Porcentaje de Colores en Español')
plt.grid(True)
plt.show()
```
