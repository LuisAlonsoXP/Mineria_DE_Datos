```python
import pandas as pd

df = pd.read_csv('items.csv')
df.eval('importe = precio * cantidad', inplace=True)
dfa = df.query("categoria == 'A'")
promedio = df.groupby('categoria')['importe'].mean()
print(promedio)
```

```python
import pandas as pd

dfc = pd.read_csv('clientes.csv')
dfo = pd.read_csv('ordenes.csv')
merged = dfc.merge(dfo, on='cliente_id', how='left')
print(len(merged))
monto_por_cliente = merged.groupby('cliente_id')['monto'].sum().fillna(0)
print(monto_por_cliente)
```

```python
import pandas as pd

df = pd.read_csv('sensores.csv', na_values=['NA', ''])
df = df[['ts', 'temperatura', 'humedad']]
df['ts'] = pd.to_datetime(df['ts'], format='mixed', errors='coerce')
df['temperatura'] = df['temperatura'].fillna(df['temperatura'].median())
df['humedad'] = df['humedad'].fillna(df['humedad'].mean())
print(df['ts'].isna().sum())
print(round(df['temperatura'].median(), 1))
print(round(df['humedad'].mean(), 3))
```
